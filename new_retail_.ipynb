```python
import seaborn as sns#Sets up a clean chart space of a specific size
import matplotlib.pyplot as plt#Sets up a clean chart space of a specific size
plt.figure(figsize=(5, 4))#Creates a new figure (canvas) for the plot.
sns.barplot(x='Country', y='Total_Amount', data=a,hue='Income',color='purple')#creating chart
plt.title('country amount')#title
plt.xlabel('country')#xlable title
plt.ylabel('Total amount')#y lable title
plt.show()#show
```

```python
plt.figure(figsize=(10,5))# Sets a wide chart size for better readability
sns.lineplot(x='Product_Type', y='Total_Purchases',data=a,color='blue',hue='Gender')# makes the bars stand out in red
plt.title('which product purchases most')#Adds a title
plt.xlabel('product type')#xlable name
plt.ylabel('total purchases')#ylable
plt.xticks(rotation=90)#to rotate x axis lable
plt.grid(True)#Adds grid lines for easier value comparison
plt.show()#show chart
```

```python
plt.figure(figsize=(10, 6))# Set the figure size to 10 inches wide and 6 inches tall
plt.hist(a['Total_Amount'], bins=20, color='purple')# Create a histogram for the 'Manufacturing costs' column with 20 bins and green bars
plt.title('frequency of total amount')# Set the title of the plot
plt.xlabel('total smount')# Label the X-axis
plt.ylabel('Frequency')# Label the Y-axis (frequency of data points in each bin)
plt.grid(axis='y')# Display horizontal grid lines for better readability (only along Y-axis)
plt.show()# Show the plot
```

```python
plt.figure(figsize=(8,5))#Sets the size of the chart
Product_Brand = a['Product_Brand'].value_counts()
plt.pie(Product_Brand, labels=Product_Brand.index, autopct='%1.1f%%', colors=['violet', 'blue', 'green', 'orange', 'red', 'purple', 'brown', 'pink', 'gray', 'olive'])
plt.title('Product Brand Distribution')#Adds a descriptive title to the chart
plt.show()#show
```

```python
heatmap_data = a[['Total_Purchases', 'Amount','Total_Amount']]# Select only the relevant columns for correlation analysis
correlation_matrix_subset = heatmap_data.corr()# Calculate the correlation matrix for the selected columns
plt.figure(figsize=(6, 4))# Set the figure size to 6 inches wide and 4 inches tall
sns.heatmap(correlation_matrix_subset, annot=True, cmap='coolwarm', fmt=".2f")# Create a heatmap of the correlation matrix
plt.title('Total_Purchases:Amount vs.Total_Amount')# Set the title of the heatmap
plt.show()# Display the plot
```

```python
plt.figure(figsize=(12, 6))# Set the figure size to 12 inches wide and 6 inches tall
sns.boxplot(x='Product_Category',y='Total_Amount',data=a)# Create a boxplot to show the distribution of Revenue generated across different Customer demographics
plt.xlabel('Product_Category')# Label the X-axis
plt.ylabel('Total_Amount')# Label the Y-axis
plt.title('total amount by Product_Category')# Set the title of the plot
plt.show()# Display the plot
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create subplot grid
fig, axes = plt.subplots(nrows=4, ncols=2, figsize=(15, 25))
axes = axes.flatten()

# Plot 1: Country vs Total Amount (barplot with Income hue)
sns.barplot(x='Country', y='Total_Amount', data=a, hue='Income', color='purple', ax=axes[0])
axes[0].set_title('Country Amount')
axes[0].set_xlabel('Country')
axes[0].set_ylabel('Total Amount')

# Plot 2: Product Category vs Age (lineplot)
sns.lineplot(x='Product_Category', y='Age', data=a, color='blue', ax=axes[1])
axes[1].set_title('Product Category by Age')
axes[1].set_xlabel('Product Category')
axes[1].set_ylabel('Age')
axes[1].grid(True)

# Plot 3: Product Type vs Purchases (by Gender)
sns.lineplot(x='Product_Type', y='Total_Purchases', data=a, hue='Gender', ax=axes[2])
axes[2].set_title('Which Product Purchases Most')
axes[2].set_xlabel('Product Type')
axes[2].set_ylabel('Total Purchases')
axes[2].tick_params(axis='x', rotation=90)
axes[2].grid(True)

# Plot 4: Histogram - Total Amount
axes[3].hist(a['Total_Amount'], bins=20, color='purple')
axes[3].set_title('Frequency of Total Amount')
axes[3].set_xlabel('Total Amount')
axes[3].set_ylabel('Frequency')
axes[3].grid(axis='y')

# Plot 5: Product Brand Distribution (Pie)
Product_Brand = a['Product_Brand'].value_counts()
axes[4].pie(Product_Brand, labels=Product_Brand.index, autopct='%1.1f%%',
            colors=['violet', 'blue', 'green', 'orange', 'red', 'purple',
                    'brown', 'pink', 'gray', 'olive'])
axes[4].set_title('Product Brand Distribution')

# Plot 6: Heatmap - Correlation
heatmap_data = a[['Total_Purchases', 'Amount', 'Total_Amount']]
correlation_matrix_subset = heatmap_data.corr()
sns.heatmap(correlation_matrix_subset, annot=True, cmap='coolwarm', fmt=".2f", ax=axes[5])
axes[5].set_title('Correlation: Purchases, Amount, Total Amount')

# Plot 7: Order Status vs Ratings (by Shipping Method)
sns.barplot(x='Order_Status', y='Ratings', data=a, hue='Shipping_Method', ax=axes[6])
axes[6].set_title('Ratings by Order Status & Shipping Method')
axes[6].set_xlabel('Order Status')
axes[6].set_ylabel('Ratings')
axes[6].tick_params(axis='x', rotation=90)
axes[6].grid(True)

# Plot 8: Boxplot - Product Category vs Total Amount
sns.boxplot(x='Product_Category', y='Total_Amount', data=a, ax=axes[7])
axes[7].set_title('Total Amount by Product Category')
axes[7].set_xlabel('Product Category')
axes[7].set_ylabel('Total Amount')



plt.tight_layout()
plt.show()
```

1. Country vs Total Amount (by Income)

Certain countries contribute a higher total purchase value.

Within countries, higher income groups usually spend more, but you may notice some countries where lower-income customers are still buying frequently.

Useful for regional targeting + income-based segmentation.

2. Product Category vs Age

Some categories are popular in specific age groups (e.g., younger customers buying fashion/electronics, older age groups preferring household/health).

Helps in personalized product recommendations by age.

3. Product Type vs Purchases (by Gender)

Certain product types show clear gender preference.

For example, women may dominate fashion & beauty purchases, while men may lean toward electronics or sports.

Useful for gender-focused marketing campaigns.

4. Histogram – Total Amount

Distribution shows whether most purchases are low-value frequent orders or few high-value orders.

If skewed, company can decide whether to push premium offers or increase frequency of small orders.

5. Product Brand Distribution (Pie Chart)

A few brands dominate the sales.

Identifies top-performing brands and potential partnerships.

Smaller share brands could need discounts or promotions.

6. Correlation Heatmap

Strong correlation between Total Purchases and Total Amount is expected.

If Amount vs Total_Purchases is weak, it means some customers buy fewer items but high-value products.

Highlights revenue drivers (quantity vs premium products).

7. Order Status vs Ratings (by Shipping Method)

Certain shipping methods may have higher ratings.

Cancelled/returned orders likely have lower satisfaction.

Can guide logistics improvements.

8. Product Category vs Total Amount (Boxplot)

Reveals outliers: some categories may have occasional very high-value purchases.

Median values show which categories contribute steady revenue.

Helps in inventory and pricing strategy.
